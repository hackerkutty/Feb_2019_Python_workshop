# Visualization Plots using matplotlib/Seaborn libraries



```python
# Importing all the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pylab as plt
import seaborn as sns

%matplotlib inline
seaborn.set()
```

```python
# Reading the csv file using pandas 
a=pd.read_csv('Customer_churn_prediction.csv')

```

# Histogram plot using matplotlib

```python
plt.hist(a.ApplicantIncome)
plt.show()
```

# Scatter plot using matplotlib

```python

# Create sample data, add some noise
x = np.random.uniform(1, 100, 1000)
y = np.log(x)

plt.scatter(x, y)
plt.show()
```

# Scatter plot using matplotlib adding some noise to y

```python
# Create sample data, add some noise
x = np.random.uniform(1, 100, 1000)
y = np.log(x) + np.random.normal(0, .3, 1000)

plt.scatter(x, y)
plt.show()
```

# Scatter plot between Income and LoanAmount

```python
plt.scatter(a.ApplicantIncome, a.LoanAmount, alpha=0.5)
#plt.scatter(a.ApplicantIncome, a.LoanAmount, c='colors',alpha=0.5)
plt.show()
```

```python
# Parameterzing the colour 
plt.scatter(a.ApplicantIncome, a.LoanAmount, c='red', alpha=0.5)
plt.show()
```

# Scatter plot between x,y and area as parameter

```python
# Fixing random state for reproducibility
sns.set()
np.random.seed(100) # Specificy the importance of random seed


N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
plt.scatter(x, y, alpha=0.5)
#plt.scatter(x, y, c='colors', alpha=0.5)
plt.show()
```
